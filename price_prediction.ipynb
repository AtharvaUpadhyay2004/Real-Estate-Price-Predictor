## USING THE SCIKIT LEARN MODULE, CONDUCTING TRAIN TEST SPLITTING

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("data.csv")
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data, test_size = 0.2, random_state = 42)
print(f"Rows in train set: {len(train_set)} \n Rows in test set: {len(test_set)}")   
```

since the train set and the test set should contain equal ratio of components of the data, if not then the code become bias with the same category of data. To avoid this we use stratified random sampling to make sure the ratio of components is same in both sets.

```python
import pandas as pd
data = pd.read_csv("data.csv")
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(data, data[' CHAS']):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]
```

## CREATING A PIPELINE

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "median")
imputer.fit(data) 
# x = imputer.transform(data)
# data_tf = pd.DataFrame(x, columns = data.columns)
#using imputer to fill missing values
from sklearn.pipeline import Pipeline    
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'median')),
    ('std_scaler', StandardScaler()),
    # can add more  
]) 
```

```python
data_num_tf = my_pipeline.fit_transform(data)
data_num_tf
```

## EVALUATING THE MODEL

```python
#using linear regression
import numpy as np
from sklearn.metrics import mean_squared_error
data_predictions = model.predict(data_num_tf)
lin_mse = mean_squared_error(data_lable,data_predictions)
lin_mse = np.sqrt(lin_mse)
lin_mse 
```

```python
#using decision tree regressor
from sklearn.metrics import mean_squared_error
data_predictions = model2.predict(data_num_tf)
mse = mean_squared_error(data_lable,data_predictions)
rmse = np.sqrt(lin_mse)
rmse 
# over fitting case
```

```python
#using decision tree regressor
from sklearn.metrics import mean_squared_error
data_predictions = model2.predict(data_num_tf)
mse = mean_squared_error(data_lable,data_predictions)
rmse = np.sqrt(lin_mse)
rmse 
# over fitting case
```

## USING BETTER EVALUATION TECHNIQUE - CROSS VALIDATION

```python
# for model (LinearRegression)
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,data_num_tf,data_lable, scoring= "neg_mean_squared_error")
rmse_scores1 = np.sqrt(-scores)
rmse_scores1
```

```python
# for model2 (Decision tree regressor)
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model2,data_num_tf,data_lable, scoring= "neg_mean_squared_error")
rmse_scores2 = np.sqrt(-scores)
rmse_scores2
```

```python
# for model3 (Random Forest regressor)
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model3,data_num_tf,data_lable, scoring= "neg_mean_squared_error")
rmse_scores3 = np.sqrt(-scores)
rmse_scores3
```

```python
def print_scores(scores):
    print("scores :", scores)
    print("the mean score is :", scores.mean())
    print("the SD of score is :", scores.std())

print_scores(rmse_scores1)
print("\n")
print_scores(rmse_scores2)
print("\n")
print_scores(rmse_scores3)
```
